```python
from apyori import apriori

assoc_results = apriori(df_into_list, min_support = 0.0035, min_confidence = 0.25, min_lift = 3, min_length = 2)
```

```python
assoc_results = list(assoc_results)
print(assoc_results[0])
```

```python
print(len(assoc_results))
```

```python
results = []
```

```python
for item in assoc_results:
    
    item_sets  = item[0]
    items = [x for x in item_sets]
    
    value0 = str(items[0])
    value1 = str(items[1]) 
    value2 = str(item[1])
    value3 = str(item[2][0][2])
    value4 = str(item[2][0][3])
    
    row = (value0, value1, value2, value3, value4)
    results.append(row)
    
labels = ['1', '2', 'Support', 'Confidence', 'Lift']
```

```python
rules = pd.DataFrame.from_records(results, columns = labels)
print(rules)
```
